```python
import numpy as np
import pandas as pd 
import datetime as dt
from bs4 import BeautifulSoup
import requests
import csv
import time
```

```python
cols=['Match_no','Home_Team','Away_Team','Home_Goals','Away_goals','Result','Player_team','Player','Type']
```

```python
def AddTeamsToURL(URL):
    URL=URL.split('/')
    URL.insert(5,'teams')
    URL='/'.join(URL)
    return URL

AddTeamsToURL('https://www.skysports.com/football/watford-vs-man-utd/408159')
```

```python
def CalculateResult(scorelist):
    if scorelist[0]>scorelist[1]:
        return 'HW'
    elif scorelist[0]==scorelist[1]:
        return 'D'
    else:
        return 'HL'
```

```python
def GetMatchData(URL,match_no,matchlist):
    URL=AddTeamsToURL(URL)
    source = requests.get(URL).text
    soup2 = BeautifulSoup(source, 'lxml')
    
    #Scrape home and away teams
    teamlist=[]
    for team in soup2.findAll('abbr',{"class": "swap-text--bp10"}):
        teamlist.append(team['title'])
        
    #Scrape home and away goals  
    scorelist=[]
    for score in soup2.findAll('span',{"class": "match-head__score"}):
        scorelist.append(int(score.text.strip()))
        
    #Calculate match results    
    result=CalculateResult(scorelist)
    
    #Scrape player numbers
    i=1
    for player in soup2.findAll('span',{"class": "team-lineups__list-player-number"}):
        plyr=player.text.strip()
        if i<=11:
            matchlist.append([match_no,teamlist[0],teamlist[1],scorelist[0],scorelist[1],result,'Home',plyr,'Starter'])
        elif i>11 and i<=18:
            matchlist.append([match_no,teamlist[0],teamlist[1],scorelist[0],scorelist[1],result,'Home',plyr,'Sub'])
        elif i>18 and i<=29:
            matchlist.append([match_no,teamlist[0],teamlist[1],scorelist[0],scorelist[1],result,'Away',plyr,'Starter'])
        else:
            matchlist.append([match_no,teamlist[0],teamlist[1],scorelist[0],scorelist[1],result,'Away',plyr,'Sub'])
        i+=1
            
    
    
```

```python
def LookForMatches(FixturesURL,matchlist):
    source = requests.get(FixturesURL).text
    soup = BeautifulSoup(source, 'lxml')
    i=1
    for match in soup.findAll('a',{"class": "matches__item matches__link"}):      
        matchURL=match['href']
        print(matchURL)
        GetMatchData(matchURL,i,matchlist)
        i+=1
        time.sleep(np.random.randint(4,high=12))

    
```

```python
matchlist=[]
LookForMatches('https://www.skysports.com/premier-league-results/2016-17',matchlist)
```

```python
AllMatchData=pd.DataFrame(matchlist,columns=cols)
AllMatchData.info()
```

```python
def PlayerID(Home_Team,Away_Team,Player_Team,Player):
    if Player_Team=='Home':
        return Home_Team+"#"+str(Player)
    else:
        return Away_Team+"#"+str(Player)
```

```python
AllMatchData['Player_ID']=AllMatchData.apply(lambda x:PlayerID(x['Home_Team'],x['Away_Team'],x['Player_team'],x['Player']),axis=1)
```

## Convert to format needed for machine learning

### Considering only starting XI

```python
StartingXI=CombinedData[CombinedData['Type']=='Starter']
```

```python
StartingXI_Result=StartingXI[['Match_no','Result']].drop_duplicates()
```

```python
StartingXI=StartingXI.groupby(by=['Match_no','Player_team'],as_index=False).sum()
```

```python
StartingXI=StartingXI.merge(StartingXI_Result,how='inner',on='Match_no')
```

```python
StartingXI
```

```python
StartingXI=StartingXI[['Match_no', 'Result','Player_team','overall','pace', 'shooting', 'passing', 'dribbling','defending', 'physic']]
```

```python
StartingXI_H=StartingXI[StartingXI['Player_team']=='Home']
```

```python
StartingXI_A=StartingXI[StartingXI['Player_team']=='Away']
```

```python
StartingXI_H.reset_index(drop=True,inplace=True)
StartingXI_A.reset_index(drop=True,inplace=True)
```

```python
StartingXI_Res=StartingXI_H['Result']
StartingXI_H.drop(columns=['Player_team','Match_no','Result'],inplace=True)
StartingXI_A.drop(columns=['Player_team','Match_no','Result'],inplace=True)
```

```python
StartingXI_H.info()
```

```python
Match_Qual_Diff=StartingXI_H.subtract(StartingXI_A)

Match_Qual_Diff.head()
```

```python
Match_Qual_Diff['Results']=StartingXI_Res
```
